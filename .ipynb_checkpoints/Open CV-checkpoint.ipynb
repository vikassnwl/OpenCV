# Basic image enhancement using mathematical operations

## Addition or Brightness

```python
matrix = np.ones(image.shape, dtype='uint8') * 50

image_brighter = cv2.add(image, matrix)
image_darker = cv2.subtract(image, matrix)

plt.figure(figsize=[21,7])

plt.subplot(141);plt.imshow(image_darker);plt.title('Darker');
plt.subplot(142);plt.imshow(image);plt.title('Original');
plt.subplot(143);plt.imshow(image_brighter);plt.title('Brighter');
```

## Multiplication or Contrast

```python
matrix1 = np.ones(image.shape) * .8
matrix2 = np.ones(image.shape) * 1.5

image_darker = np.uint8(cv2.multiply(np.float64(image), matrix1))
image_brighter = np.uint8(cv2.multiply(np.float64(image), matrix2))

# Show the images
plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(image_darker);plt.title('Lower Contrast');
plt.subplot(132);plt.imshow(image);plt.title('Original');
plt.subplot(133);plt.imshow(image_brighter);plt.title('Higher Contrast');
```

## Handling Overflow using np.clip

```python
matrix1 = np.ones(image.shape) * .8
matrix2 = np.ones(image.shape) * 1.5

image_lower = np.uint8(cv2.multiply(np.float64(image), matrix1))
image_higher = np.uint8(np.clip(cv2.multiply(np.float64(image), matrix2),0,255))

# Show the images
plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(image_lower);plt.title('Lower Contrast');
plt.subplot(132);plt.imshow(image);plt.title('Original');
plt.subplot(133);plt.imshow(image_higher);plt.title('Higher Contrast');
```

# Application: Logo Manipulation

## Read foreground image

```python
logo = cv2.imread('coca_cola_logo.png')
logo.shape
```

```python
plt.imshow(logo[:,:,::-1])
logo_h, logo_w = logo.shape[:2]
```

## Read background image

```python
bg_img = cv2.imread('checkerboard_colored.png')

# Set desired width (logo_w) and maintain image aspect ratio
aspect_ratio = logo_w/bg_img.shape[1]
dim = (logo_w, int(bg_img.shape[0]*aspect_ratio))

# Resize background image to same size as logo image
bg_img = cv2.resize(bg_img, dim, interpolation=cv2.INTER_AREA)

plt.imshow(bg_img)
bg_img.shape
```

## Create mask for original image

```python
img_gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)

#Apply global thresholding to create a binary mask of the logo
retval, img_mask = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)

plt.imshow(img_mask, cmap='gray')
img_mask.shape
```

## Invert the mask

```python
# Create an inverse mask
img_mask_inv = cv2.bitwise_not(img_mask)
plt.imshow(img_mask_inv, cmap='gray');
```

## Apply background on the mask

```python
# Create colorful background "behind" the logo lettering
bg_img = cv2.bitwise_and(bg_img, bg_img, mask=img_mask)
plt.imshow(bg_img[:,:,::-1]);
```

## Isolate foreground from image

```python
# Isolate foreground (red from original image) using the inverse mask
fg_img = cv2.bitwise_and(logo, logo, mask=img_mask_inv)
plt.imshow(fg_img[:,:,::-1]);
```

## Result: Merge foreground and background

```python
# Add the two previous results obtain the final result
result = cv2.add(bg_img[:,:,::-1], fg_img[:,:,::-1])
plt.imshow(result);
```

# Extracting rose from image

```python
# reading rose image
rose_bgr = cv2.imread('rose.jpg')
rose_rgb = rose_bgr[:,:,::-1]
```

```python
plt.imshow(rose_rgb);
```

```python
# bgr2hsv
rose_hsv = cv2.cvtColor(rose_bgr, cv2.COLOR_BGR2HSV)
```

```python
# using inRange ot extract red color
# lower boundary RED color range values; Hue (0 - 10)
lower1 = np.array([0, 100, 20])
upper1 = np.array([10, 255, 255])
 
# upper boundary RED color range values; Hue (160 - 180)
lower2 = np.array([160,100,20])
upper2 = np.array([179,255,255])
 
lower_mask = cv2.inRange(rose_hsv, lower1, upper1)
upper_mask = cv2.inRange(rose_hsv, lower2, upper2)
 
full_mask = lower_mask + upper_mask;
```

```python
plt.imshow(full_mask, cmap='gray');
```

```python
result = cv2.bitwise_and(rose_rgb, rose_rgb, mask=full_mask)
plt.imshow(result);
```

# Image Transformation

```python
rose_bgr = cv2.imread('rose.jpg')
cols = rose_bgr.shape[1]
rows = rose_bgr.shape[0]
M = np.float32([[1, 0, -150], [0, 1, -70]])
shifted = cv2.warpAffine(rose_bgr, M, (cols, rows))
plt.imshow(shifted[:,:,::-1]);
```

# Image Rotation

```python
img_bgr = cv2.imread('scenery.jpeg')
rows, cols = img_bgr.shape[:2]
center = cols/2, rows/2
angle1 = -90
angle2 = 90
M1 = cv2.getRotationMatrix2D(center, angle1, 1)
M2 = cv2.getRotationMatrix2D(center, angle2, 1)
rotateneg90 = cv2.warpAffine(img_bgr, M1, (cols, rows))
rotate90 = cv2.warpAffine(img_bgr, M2, (cols, rows))
plt.figure(figsize=[15,5])
plt.subplot(131);plt.imshow(img_bgr[:,:,::-1]);plt.title('Original');
plt.subplot(132);plt.imshow(rotateneg90[:,:,::-1]);plt.title('Rotated -90');
plt.subplot(133);plt.imshow(rotate90[:,:,::-1]);plt.title('Rotated 90');
```
